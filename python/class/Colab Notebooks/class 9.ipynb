```
# append - It add a value at the end of the list

# Defining the List
list_demo = []
print(f"Initially the list is - {list_demo}")
limit = int(input())
for _ in range(limit):
  # print(f"List Now - {list_demo}")
  val = input('Enter some value:- ')
  list_demo.append(val)
print(f"The length of list is - {len(list_demo)}")
print(f"Finally the list is - {list_demo}")
```

```
vals = [1,2,3]
print(vals)
vals.append([5, 4])
print(vals)
```

```
## extend - It can add more that one values to the list
print(list_demo)
print(len(list_demo))
list_demo.extend([1,2,3,4,5])
print(list_demo)
print(len(list_demo))
```

```
## insert - Add values to the specified index location in the list
print(list_demo)
print(len(list_demo))
location = int(input())
new_vals = ['Saanvi', 'Subhadeep']
list_demo.insert(location, new_vals)
print(list_demo)
print(len(list_demo))
```

```
list_demo.insert(5, 'FRIENDS')
print(list_demo)
```

```
## index - Tells us the location of the specified value in the list
list_demo.index(['Saanvi', 'Subhadeep'])
```

```
## count - It tells us the occurence of a specified value in the list
nu_list = [1,1,1,2,1,1,2,2,3,4,5,1,1]
print(nu_list.count(1))
print(nu_list.count(2))
```

```
## clear - It wipes out the data form the whole list
print(list_demo)
print(list_demo.clear())
print(list_demo)
```

```
## reverse - It reverses the order of the values in the list
list_demo = [1,2,3,4,5]
print(list_demo)
list_demo.reverse()
print(list_demo)
```

```
## pop - It pops the value from the end of the list and returns it
print(list_demo)
print(list_demo.pop())
print(list_demo)
```

```
#removes everything from a list
print(list_demo)
list_demo.clear()
list_demo = [9,18,1,84,848,4,81,4,98,94,984,984,94,1]
```

```
## sort- returns the list in ascending or descending order
print("The most unsorter list ever")
print(list_demo)
print('----------------------------------------')
list_demo.sort(reverse=True)
print(list_demo)
print(list_demo)
```

```
## sorted- numerical order ascending or descending. returns a copy of the list
print("The most unsorter list ever")
print(list_demo)
print('----------------------------------------')
print(sorted(list_demo))
print(list_demo)
print(list_demo)
```

```
## remove- removes the first occurence of the specified value
list_demo = [1,2,3,4,94,2,1,49,2,62,9]
for i in list_demo:
  print(f"{i} --> {list_demo.count(i)}")
```

```
demo = [1,2,3,4,1,2,5]
print(demo)
demo.remove(2)
print(demo)
```

```
# copy- returns a copy of the list
print(demo)
new_list = []
print(new_list)
new_list = demo.copy()
print(new_list)
```

```
## del- wipes everything
print(demo)
demo.clear()
print(demo)
del demo
# print(demo)
```
