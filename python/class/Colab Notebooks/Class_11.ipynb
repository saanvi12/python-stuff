```python
class Class1:
  def __init__(self, fname, lname, age, add):
    self.fname = fname
    self.lname = lname 
    # self.age = int(input("Enter your Age :- "))
    # self.add = input("Enter your Address :- ")
    self.age = age
    self.add = add

  def addToFile(self):
    with open('database.txt', 'a') as f:
      f.write(f"{self.fname}, {self.lname}, {self.age}, {self.add}")
      f.write("\n")
```

```python
values = [
          ["saanvi", "Mogla", 14, "asdasd"],
          ["Subhadeep", "Banejee", 21, "asdasd"],
          ["Richard", "Chakraborty", 20, "asdasd"],
          ["Edison", "ching", 27, "asdasd"],
          ["Gandalf", "The Great", 114, "asdasd"],
]

for i in values:
  obj = Class1(i[0], i[1], i[2], i[3])
  obj.addToFile()
```
