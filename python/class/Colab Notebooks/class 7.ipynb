```
# #a. Below 25 - F
# b. 25 to 45 - E
# c. 45 to 50 - D
# d. 50 to 60 - C
# e. 60 to 80 - B
# f. Above 80 - A
mark = float(input("Enter your mark here: "))
if mark<25:
  print("You got an F. Try better next time!")
elif mark>25 and mark <45:
  print("You got an E. You can do better, you got this!")
elif mark>45 and mark<50:
  print("You got a D. You're getting better!")
elif mark>50 and mark<60:
  print("You got a C. Not bad!")
elif mark>60 and mark<80:
  print("You got a B. Nice!")
else:
  print("You got an A! Great job!")
```

```
length = int(input("Enter the length here: "))
height = int(input("Enter the height here: "))
area = length*height
if length==height:
  print(f"This is a square and the area is {area}!")
else:
  print(f"This is a rectangle and the area is {area}!")
```

```
salary = float(input("Enter your salary here: "))
years = float(input("Enter the number of years you've worked with us here: "))
if years>5:
  print("You are eligible for a bonus of 5%!")
  print((5*salary)/100)
  print("The above number is your additional bonus. Thank you for your dedication!")
else:
  print("Sorry, but you are not eligible for a bonus.")
  print(5-years)
  print("The above number is the number of years in which you will be eligible for a bonus!")
```
